## 1) Lists: Multiple readings over time
Lists are ordered collections. They’re great for time-ordered health data like a series of BP or glucose readings.

```python
# Example: systolic blood pressure readings taken across several days
systolic_readings = [118, 122, 135, 142, 128, 150]

print("All systolic readings:", systolic_readings)
print("First reading:", systolic_readings[0])
print("Most recent reading:", systolic_readings[-1])

# Basic summary ideas (we'll do more of this later with pandas)
print("Number of readings:", len(systolic_readings))
print("Minimum:", min(systolic_readings))
print("Maximum:", max(systolic_readings))
print("Average:", sum(systolic_readings) / len(systolic_readings))
```

### Common list operations in health informatics
- Add a new reading (new measurement arrives)
- Remove an incorrect reading (data cleaning)
- Sort readings (ensure time order)


```python
systolic_readings = [118, 122, 135, 142]
print("Starting list:", systolic_readings)

# Add a new reading
systolic_readings.append(128)
print("After append:", systolic_readings)

# Remove a reading (example: data entry error)
systolic_readings.remove(122)
print("After remove:", systolic_readings)

# Sort (useful if data arrived out of order)
systolic_readings.sort()
print("After sort:", systolic_readings)
```

## 2) Tuples: Fixed records (immutable)
Tuples are ordered like lists, but you typically use them for **fixed** groupings that shouldn't be changed—like a single vital measurement record.

```python
# A single BP record: (date, systolic, diastolic)
bp_record = ("2026-03-05", 128, 82)

print("BP record tuple:", bp_record)
print("Date:", bp_record[0])
print("Systolic:", bp_record[1])
print("Diastolic:", bp_record[2])

# Tuples are often used inside lists to represent a series of records:
bp_series = [
    ("2026-03-01", 135, 88),
    ("2026-03-03", 142, 90),
    ("2026-03-05", 128, 82),
]

print("\nBP series (list of tuples):")
for rec in bp_series:
    print(rec)
```

### Why tuples matter
In health informatics, you’ll often see a “row” of data represented like a tuple:
- (patient_id, date, lab_value)
- (encounter_id, code, description)

Later, pandas DataFrames behave similarly—rows are fixed records.


## 3) Dictionaries: Patient records (key-value)
Dictionaries map **keys** to **values**. They’re perfect for representing a patient chart, where each field has a name.

```python
patient = {
    "patient_id": "P10023",
    "name": "Jordan Lee",
    "age": 63,
    "sex": "F",
    "conditions": ["hypertension", "hyperlipidemia"],
    "medications": ["lisinopril", "atorvastatin"],
    "latest_bp": {"systolic": 128, "diastolic": 82},
}

print("Patient dictionary:")
print(patient)

print("\nAccess specific fields:")
print("Name:", patient["name"])
print("Conditions:", patient["conditions"])
print("Latest systolic:", patient["latest_bp"]["systolic"])
```

### Updating a dictionary (chart updates)
EHRs get updated: new meds, new diagnoses, new labs.


```python
patient = {"patient_id": "P10023", "medications": ["lisinopril"]}

# Add a new field
patient["pcp"] = "Dr. Nguyen"

# Update meds (add a medication)
patient["medications"].append("atorvastatin")

# Record a new lab result
patient["latest_a1c"] = 6.4

print(patient)
```

### Looping through dictionaries (batch processing)
You can store many patients as a list of dictionaries and process them.


```python
patients = [
    {"name": "Alicia", "age": 67, "conditions": ["hypertension"]},
    {"name": "Ben", "age": 52, "conditions": ["asthma"]},
    {"name": "Carlos", "age": 74, "conditions": ["hypertension", "diabetes"]},
]

for p in patients:
    if "diabetes" in p["conditions"]:
        print(p["name"], "→ diabetes registry candidate")
```

## 4) Sets: Unique diagnoses or medications
Sets store **unique** values (no duplicates). They are useful for counting unique items and finding overlap between lists (e.g., shared conditions).

```python
# Example: diagnoses captured across multiple encounters (duplicates happen!)
diagnosis_codes = ["I10", "E11.9", "I10", "E78.5", "E11.9"]

unique_dx = set(diagnosis_codes)

print("All codes:", diagnosis_codes)
print("Unique codes:", unique_dx)
print("Number of unique diagnoses:", len(unique_dx))
```

### Set operations: Overlap between medication lists
This is useful for identifying:
- Shared meds between two formularies
- Overlap between two patients (potential interactions, duplication)


```python
patient_a_meds = {"lisinopril", "atorvastatin", "metformin"}
patient_b_meds = {"metformin", "amlodipine", "aspirin"}

print("Overlap (intersection):", patient_a_meds & patient_b_meds)
print("All meds combined (union):", patient_a_meds | patient_b_meds)
print("In A but not B (difference):", patient_a_meds - patient_b_meds)
```

## 5) Choosing the right data structure (quick guide)

- **List**: ordered collection, duplicates allowed  
  *Example:* readings over time: `[118, 122, 135]`

- **Tuple**: ordered, fixed record  
  *Example:* one measurement: `("2026-03-05", 128, 82)`

- **Dictionary**: structured record with named fields  
  *Example:* patient chart: `{"name": "...", "age": 63, ...}`

- **Set**: unique items, good for overlap and de-duplication  
  *Example:* unique diagnosis codes: `{"I10", "E11.9"}`
