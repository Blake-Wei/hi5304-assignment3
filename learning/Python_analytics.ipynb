One sample t-test

```python
import numpy as np
from scipy import stats
# Sample data
data = [5.1, 4.8, 5.3, 5.7, 5.2, 4.9, 5.4, 5.6, 5.0, 5.3]
# Hypothesized population mean
mu = 5.0
# Perform one-sample t-test
t_stat, p_value = stats.ttest_1samp(data, mu)
# Print the result
print(f'T-statistic: {t_stat}')
print(f'P-value: {p_value}')
```

Independent sample t-test

```python
import numpy as np
from scipy import stats
# Sample data
group1 = [5.1, 4.8, 5.3, 5.7, 5.2, 4.9, 5.4, 5.6, 5.0, 5.3]
group2 = [4.7, 4.6, 4.8, 4.9, 5.1, 4.5, 4.6, 4.7, 4.8, 4.9]
# Perform independent samples t-test
t_stat, p_value = stats.ttest_ind(group1, group2)
# Print the result
print(f'T-statistic: {t_stat}')
print(f'P-value: {p_value}')
```

One-way ANOVA

```python
import pandas as pd
from scipy.stats import f_oneway

# Step 1: Prepare the Data
data = {
    'AgeGroup': ['18-29', '18-29', '18-29', '30-44', '30-44', '30-44', 
                 '45-59', '45-59', '45-59', '60+', '60+', '60+'],
    'BMI': [22.5, 24.1, 23.0, 27.5, 26.3, 28.1, 29.2, 30.5, 28.9, 27.0, 26.4, 25.9]
}

df = pd.DataFrame(data)

# Step 2: Split the data by AgeGroup
grouped_data = df.groupby('AgeGroup')['BMI'].apply(list)

# Step 3: Run the One-Way ANOVA
anova_result = f_oneway(*grouped_data)

# Step 4: Output the result
print(f"F-statistic: {anova_result.statistic}")
print(f"p-value: {anova_result.pvalue}")
```

```python
# Logistic regression

import numpy as np
import pandas as pd
import statsmodels.api as sm
```

```python
np.random.seed(42)

n = 100

# Predictor
bmi = np.random.normal(loc=27, scale=4, size=n)

# Logistic model (true relationship)
log_odds = -10 + 0.35 * bmi
prob = 1 / (1 + np.exp(-log_odds))

# Binary outcome
hypertension = np.random.binomial(1, prob)

# Create DataFrame
df = pd.DataFrame({
    'bmi': bmi,
    'hypertension': hypertension
})

df.head()
```

```python
X = df[['bmi']]
X = sm.add_constant(X)   # adds intercept
y = df['hypertension']

model = sm.Logit(y, X)
result = model.fit()

result.summary()
```

```python
# Linear regression

import pandas as pd
import statsmodels.api as sm
# Step 1: Prepare the Data
data = pd.DataFrame({
'Subject': range(1, 11),
'PhysicalActivity': [5, 3, 7, 4, 6, 2, 8, 4, 5, 6],
'BMI': [22.5, 24.1, 23.0, 27.5, 26.3, 28.1, 29.2, 30.5, 28.9, 27.0]})
# Step 2: Define the independent variable (PhysicalActivity) and the dependent variable (BMI)
X = data['PhysicalActivity']
y = data['BMI']
# Add a constant to the independent variable (for the intercept)
X = sm.add_constant(X)
# Step 3: Run the Linear Regression
model = sm.OLS(y, X).fit()
# Step 4: Interpret the Results
summary = model.summary()
# Output the result
print(summary)
```

```python
# Chi Square test of independence

import pandas as pd
from scipy.stats import chi2_contingency

# Step 1: Prepare the Data
# Create a contingency table
data = pd.DataFrame({
    'Hypertension Yes': [30, 20],
    'Hypertension No': [10, 40]
}, index=['Smoker Yes', 'Smoker No'])

# Step 2: Run the Chi-square Test
chi2, p, dof, expected = chi2_contingency(data)

# Output the results
print(f"Chi-squared (X²): {chi2}")
print(f"p-value: {p}")
print(f"Degrees of freedom (df): {dof}")
print("Expected frequencies:")
print(expected)
```

```python
# Chi Square goodness of fit test

import numpy as np
from scipy.stats import chi2

# Step 1: Prepare the Data
observed_frequencies = np.array([20, 22, 25, 18, 24, 17, 19])

# Step 2: Specify the Expected Frequencies
# For a uniform distribution, the expected frequency for each day is the total number of observations divided by the number of days
total_observations = np.sum(observed_frequencies)
num_days = len(observed_frequencies)
expected_frequencies = np.repeat(total_observations / num_days, num_days)

# Step 3: Run the Chi-square Goodness of Fit Test
chi_square_statistic = ((observed_frequencies - expected_frequencies)**2 / expected_frequencies).sum()

# Degrees of freedom
dof = num_days - 1

# Calculate p-value
p_value = chi2.sf(chi_square_statistic, dof)

chi_square_statistic, p_value
```

```python
# Mann-Whitney U test

import pandas as pd
from scipy.stats import mannwhitneyu

# Step 1: Prepare the Data
data = pd.DataFrame({
    'Subject': range(1, 11),
    'DietGroup': ['Low-Fat', 'Low-Fat', 'Low-Fat', 'Low-Fat', 'Low-Fat', 
                  'Low-Carb', 'Low-Carb', 'Low-Carb', 'Low-Carb', 'Low-Carb'],
    'Cholesterol': [180, 185, 178, 182, 177, 160, 162, 159, 165, 163]
})

# Step 2: Split the data by DietGroup
low_fat = data[data['DietGroup'] == 'Low-Fat']['Cholesterol']
low_carb = data[data['DietGroup'] == 'Low-Carb']['Cholesterol']

# Step 3: Run the Mann-Whitney U Test
mann_whitney_test = mannwhitneyu(low_fat, low_carb, alternative='two-sided')

# Output the results
print(f"Mann-Whitney U statistic: {mann_whitney_test.statistic}")
print(f"p-value: {mann_whitney_test.pvalue}")
```

```python
# Kruskal-Wallis test

import pandas as pd
from scipy.stats import kruskal

# Step 1: Prepare the Data
data = pd.DataFrame({
    'Subject': range(1, 10),
    'DietGroup': ['Control', 'Control', 'Control', 
                  'Low-Fat', 'Low-Fat', 'Low-Fat', 
                  'Low-Carb', 'Low-Carb', 'Low-Carb'],
    'Cholesterol': [200, 195, 198, 180, 185, 178, 160, 162, 159]
})

# Step 2: Group the data by DietGroup
grouped_data = data.groupby('DietGroup')['Cholesterol'].apply(list)

# Step 3: Run the Kruskal-Wallis Test
kruskal_test = kruskal(*grouped_data)

# Output the results
print(f"Kruskal-Wallis statistic: {kruskal_test.statistic}")
print(f"p-value: {kruskal_test.pvalue}")
```

```python
def categorize_blood_pressure(systolic, diastolic):
    if systolic < 120 and diastolic < 80:
        category = "Normal"
    elif 120 <= systolic < 130 and diastolic < 80:
        category = "Elevated"
    elif (130 <= systolic < 140) or (80 <= diastolic < 90):
        category = "Hypertension Stage 1"
    elif (140 <= systolic) or (90 <= diastolic):
        category = "Hypertension Stage 2"
    else:
        category = "Hypertensive Crisis (consult your doctor immediately)"
    return category

# Example usage:
systolic_bp = 135
diastolic_bp = 85
bp_category = categorize_blood_pressure(systolic_bp, diastolic_bp)
print(f"Blood Pressure Category: {bp_category}")
```
