```python
class Patient:
  def __init__(self, name, age, systolic):
    self.name = name
    self.age = age
    self.systolic = systolic
  
  def is_hypertensive(self):
    return self.systolic >= 130

p1 = Patient("Smith", 62, 145)
print(p1.is_hypertensive())
```

```python
def calculate_bmi(weight, height):
  return weight / (height ** 2)

bmi = calculate_bmi(80, 1.75)
print("BMI:", bmi)
```
