## 2) `if / elif / else`: Classifying Blood Pressure  
We can use thresholds (rules) to categorize blood pressure. This is a simplified example for teaching.

### Concept
- `if` checks the first condition
- `elif` checks the next condition *only if* previous ones were false
- `else` catches everything not matched above


```python
systolic = 115
diastolic = 62

if systolic < 120 and diastolic < 80:
    category = "Normal"
elif systolic < 130 and diastolic < 80:
    category = "Elevated"
elif systolic < 140 or diastolic < 90:
    category = "Hypertension Stage 1"
else:
    category = "Hypertension Stage 2"

print("BP:", systolic, "/", diastolic, "→", category)
```

## 3) `if / else`: Medication Adherence  
Medication adherence is often summarized as a **rate** (percent taken).  
Here we flag whether adherence meets a common threshold (80%).

```python
days_taken = 26
days_prescribed = 30

adherence_rate = days_taken / days_prescribed

print("Adherence rate:", round(adherence_rate, 2))

if adherence_rate >= 0.80:
    print("Adherence is acceptable ✅")
else:
    print("Adherence is low — follow-up recommended ⚠️")
```

## 4) `for` loops: Processing Multiple Readings  
A `for` loop repeats a block of code once for each item in a sequence (like a list).

### Why this matters in health informatics
- A patient has many readings over time
- A clinic has many patients
- A remote monitoring program may stream many data points daily


```python
systolic_readings = [118, 122, 135, 142, 128, 150]

for reading in systolic_readings:
    if reading < 120:
        category = "Normal"
    elif reading < 130:
        category = "Elevated"
    elif reading < 140:
        category = "Stage 1 Hypertension"
    else:
        category = "Stage 2 Hypertension"

    print("Reading:", reading, "→", category)
```

## 5) `for` loops with dictionaries: Flagging Higher-Risk Patients  
In health informatics, we often represent patient records using **dictionaries** (key-value pairs).

Below, each patient is represented as a dictionary with fields like `name` and `age`.


```python
patients = [
    {"name": "Alice", "age": 67},
    {"name": "Bob", "age": 52},
    {"name": "Carlos", "age": 74},
    {"name": "Diana", "age": 45}
]

for patient in patients:
    if patient["age"] >= 65:
        print(patient["name"], "is in a higher cardiovascular risk age group (65+)")
```

## 6) `while` loops: Monitoring Until a Condition is Met  
A `while` loop repeats **as long as** a condition remains true.

### When this appears in practice
- Simulation (e.g., what happens after each intervention step)
- Repeating reminders until a goal is achieved
- Quality improvement loops (repeat until metric improves)


```python
systolic = 155

while systolic > 130:
    print("Current systolic BP:", systolic, "→ recommend lifestyle intervention")
    systolic -= 5  # pretend systolic improves a bit each cycle

print("Goal reached. Current systolic BP:", systolic)
```

### Important safety note  
A `while` loop can run forever if the condition never becomes false.  
To prevent that, you can include a **counter** or use `break` when needed.


```python
systolic = 155
max_steps = 10
step = 0

while systolic > 130 and step < max_steps:
    print("Step", step + 1, "| systolic:", systolic)
    systolic -= 3
    step += 1

if systolic <= 130:
    print("Controlled ✅ final systolic:", systolic)
else:
    print("Not controlled yet ⚠️ final systolic:", systolic, "| steps used:", step)
```

## 7) Health Informatics Example: Glucose Screening  
Here’s a simple screening rule for glucose readings:
- `< 100` → Normal
- `100–125` → Prediabetes
- `>= 126` → Diabetes range

*(This is a simplified educational example.)*


```python
glucose_readings = [95, 110, 140, 180, 102, 210]

for glucose in glucose_readings:
    if glucose < 100:
        status = "Normal"
    elif glucose < 126:
        status = "Prediabetes"
    else:
        status = "Diabetes range"

    print("Glucose:", glucose, "→", status)
```
